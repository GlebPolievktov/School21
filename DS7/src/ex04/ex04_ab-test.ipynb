```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect("../data/checking-logs.sqlite")
```

```python
test_results = pd.read_sql(""" WITH Sub_tab1 AS (
    SELECT uid, MIN(first_commit_ts) AS first
    FROM test
    WHERE labname != 'project1'
    GROUP BY uid
),
Sub_tab2 AS (
    SELECT 
        CASE  
            WHEN s1.first < p.datetime THEN 'after' 
            ELSE 'before' 
        END AS time_per,
    (julianday(p.datetime, 'unixepoch') - julianday(s1.first)) * 24 AS time_diff
    FROM pageviews p 
    JOIN Sub_tab1 s1 ON s1.uid = p.uid
    WHERE p.uid != 'project 1 AND p.datetime IS NOT NULL'
)
SELECT time_per, AVG(time_diff) AS avg_d
FROM Sub_tab2
GROUP BY time_per""",con)
test_results
```
