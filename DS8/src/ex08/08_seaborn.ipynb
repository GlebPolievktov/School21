```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
con = sqlite3.connect("../data/checking-logs.sqlite")
```

```python
checker = pd.read_sql("SELECT uid,timestamp,numTrials FROM checker WHERE uid LIKE 'user%' AND status = 'ready' AND labname = 'project1' ",con,parse_dates='timestamp')
checker
```

```python
sns.set_theme()

sns.relplot(data=checker,x='timestamp',y='numTrials',kind="line",hue='uid',linewidth=3,height=10,aspect=1.5)

plt.gcf().set_facecolor('gray')
plt.title('Dynamic of commits per user in project1', fontsize=30)  
plt.xlabel('Timestamp', fontsize=15)
plt.ylabel('Number of Trials', fontsize=15)
plt.plot()
```
