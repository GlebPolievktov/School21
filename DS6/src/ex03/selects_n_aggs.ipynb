Load the JSON 

```python
df = pd.read_json("../ex03/auto.json")
df.set_index("CarNumber",inplace=True)
```

#2.1 Display the rows where the fines are greater than 2,100

```python
df[df['Fines'] > 2100]
```

#2.2 Display the rows where the fines are greater than 2,100 and the refund equals 2.

```python
df[(df['Refund'] == 2) & (df['Fines'] > 2100) ]
```

#2.3 Display the rows where the models are from the list: [’Focus’, ’Corolla’]

```python
df[df['Model'].isin(['Focus','Corolla'])]
```

#3 Make the aggregations with the make and model

```python
grouped = df.groupby('Make')
grouped['Fines'].median()
```

```python
grouped = df.groupby(by=['Make','Model'])
grouped['Fines'].median()
```

```python
grouped = df.groupby(by=['Make','Model'])
grouped['Fines'].count()
```

```python
grouped = df.groupby(by=['Make','Model'])
grouped['Fines'].std()
```

#4.1 Display the car numbers, grouped by the number of fines in descending order.

```python
df.reset_index()
grouped = df.groupby(by='CarNumber')
grouped['Fines'].count().sort_values(ascending=False)
```

#4.2 Select from the initial dataframe all the rows corresponding to the top-1 car number

```python
df[df.index == 'Y7689C197RUS']
```

#4.3 Display the car numbers grouped by sum of fines in descending order.

```python
df.reset_index()
grouped = df.groupby(by='CarNumber')
grouped['Fines'].sum().sort_values(ascending=False)
```

#4.4 Select all the rows corresponding to the top-1 car number from the initial dataframe

```python
df[df.index == 'X758HY197RUS']
```

#4.5 Display a table that answers the question, "Are there any car numbers connected to different models?"

```python
car_numbers = df.groupby('CarNumber')['Model'].nunique()
car_multiple = car_numbers[car_numbers > 1]

df[df.index.isin(car_multiple.index)]
```
