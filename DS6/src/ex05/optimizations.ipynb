Write a function that loops through the dataframe

```python
%%timeit -n1 -r1

list1 = list()
for i in range(0,len(df)):
    list1.append((df['Fines'].loc[i])/(df['Refund'].iloc[i] * df['Year'].iloc[i]))
df['NW Columns'] = list1

print(df)


```

using iterrows()

```python
df = df.drop('NW Columns',axis=1)
```

```python
%%timeit -n1 -r1
list2 = list()
for i,j in df.iterrows():
    list2.append((j['Fines'])/(j['Refund'] * j['Year']))
df['NW Column'] = list2
```

```python
print(df)
```

using apply() and a lambda function

```python
%%timeit -n1 -r1
df['NW Columns'] = df.apply(lambda x : x['Fines'] / (x['Year']* x['Refund']),axis=1)
```

```python
df
```

```python
df = df.drop('NW Columns',axis=1)
```

using Series objects from the dataframe

```python
%%timeit -n1 -r1
df['NW Columns'] = df['Fines'] / (df['Year']* df['Refund'])
```

```python
df
```

use the method .values

```python
%%timeit -n1 -r1
df['NW Columns'] = df['Fines'].values / (df['Year'].values * df['Refund'].values)
```

```python
df = df.reset_index()
```
