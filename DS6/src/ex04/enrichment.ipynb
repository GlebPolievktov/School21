Create a new dataframe with car numbers and their owners.

```python
sur = pd.read_json("../ex04/surname.json")
```

```python
sur.columns = sur.iloc[0]
sur = sur[1:]
```

```python
sur = sur.reset_index(drop=True)
```

```python
sur
```

```python
uniq_surname = concat_rows['CarNumber'].unique()
```

```python
uniq_surname
```

```python
np.random_state=21
surnames = pd.Series(np.random.choice(sur['NAME'],size=len(uniq_surname)))
surnames
```

```python

owners = pd.DataFrame()
```

```python
owners['CarNumber'] = uniq_surname
owners['SURNAME'] = surnames
```

```python
owners
```

Append five more observations to the fines dataframe.

```python
owners = owners[:len(owners)-20]
```

```python

owners
```

```python
data = {'CarNumber':['GJ02734U51RUS','WE378Q191RUS','LK5543161RUS'],'SURNAME':['SANCHEZ','JHONSONS','TURNER']}
add_3 = pd.DataFrame(data)
```

```python
owners = pd.concat([owners,add_3],ignore_index=True)
owners
```

Join the two dataframes

```python

car_only = pd.merge(fines,owners,on='CarNumber',how='inner')
car_only
```

```python
car_all = pd.merge(fines,owners,on='CarNumber',how='outer')
car_all
```

```python
car_left = pd.merge(fines,owners,on='CarNumber',how='left')
car_left
```

```python
car_right = pd.merge(fines,owners,on='CarNumber',how='right')
car_right
```

Create a pivot table from the fines dataframe

```python
pd.pivot_table(fines,aggfunc='sum',values='Fines',index=['Make', 'Model'], columns='Year')
```
