```python
import pandas as pd
```

#1 Create a dataframe Views

```python
df = pd.read_csv("../data/feed-views.log",sep='\t',names=['datetime','user'],parse_dates=['datetime'])
```

```python
df.info()
```

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['second'] = df['datetime'].dt.second
```

```python
df
```

```python
df.count()
```

#2 Create the new column daytime

```python
lab = ['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
bins = [0, 14399, 25199, 39599, 61199, 71999, 86399]
df['daytime'] = df['hour'] * 3600 + df['minute'] * 60 + df['second']
df.set_index('user',inplace=True)
df['daytime'] = pd.cut(df['daytime'], bins=bins, labels=lab)
```

```python
df
```
