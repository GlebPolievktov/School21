## Class Links

```python
links = Links("ml-latest-small/links.csv")
%timeit Links("ml-latest-small/links.csv")
```

```python
first_100_movie_ids = [link['movieId'] for link in links.links[:100]]
fields_to_fetch_initial = ['Director', 'Budget', 'Runtime', 'Cumulative Worldwide Gross', 'Title']
print(f"Предварительная загрузка данных для первых 100 фильмов...")
%timeit links.fetch_initial_data(first_100_movie_ids, fields_to_fetch_initial)
print("Предварительная загрузка данных завершена.")
```

#### Топ-5 режиссеров по количеству снятых ими фильмов

```python
%timeit links.top_directors(5)

print(f"Top {5} режиссеров:")
top_directors_result = links.top_directors(5)

for director, count in top_directors_result.items():
    print(f"{director}: {count}")
```

#### Топ-5 фильмов с самыми большими бюджетами

```python
%timeit links.most_expensive(5)
most_expensive_result = links.most_expensive(5)
print(f"Top {5} Most Expensive Movies:")
for movie, budget in most_expensive_result.items():
    print(f"{movie}: {budget}")
```

####  Топ-5 самых прибыльных фильмов

```python
%timeit links.most_profitable(5)

most_profitable_result = links.most_profitable(5)
print("Top 5 Most Profitable Movies:")
for movie, profit in most_profitable_result.items():
    print(f"{movie}: {profit}")
```

####  Топ-5 самых длинных фильма

```python
%timeit links.longest(5)

longest_result = links.longest(5)
print("Top 5 Longest Movies:")
for movie, runtime in longest_result.items():
    print(f"{movie}: {runtime} minutes")
```

####  Топ-5 фильмов с наибольшей стоимостью минуты экранного времени

```python
%timeit links.top_cost_per_minute(5)

top_cost_result = links.top_cost_per_minute(5)
print("Top 5 Movies by Cost per Minute:")
for movie, cost in top_cost_result.items():
    print(f"{movie}: ${cost}/minute")
```

#### Дополнительный метод: Топ-5 самых эффективных режиссеров по соотношению суммарной прибыли к суммарному бюджету всех фильмов

```python
%timeit links.director_efficiency(5)

print("Вычисляем топ-5 самых 'эффективных' режиссеров...")
top_efficiency_directors = links.director_efficiency(top_n=5, process_limit=500)

for director, efficiency in top_efficiency_directors.items():
    print(f"{director}: {efficiency}")
```
